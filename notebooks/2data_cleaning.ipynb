```python
import pandas as pd
```

```python
cleaned_data = pd.read_csv('../data/raw/ViewingActivity_raw.csv')
```

```python
cleaned_data = cleaned_data[cleaned_data['Profile Name'] == 'Hasss']
cleaned_data
```

```python
cleaned_data['Supplemental Video Type'].unique()
```

```python
cleaned_data = cleaned_data[cleaned_data['Supplemental Video Type'].isna()]
cleaned_data
```

```python
cleaned_data.drop(columns=['Profile Name','Country','Bookmark','Latest Bookmark','Supplemental Video Type','Attributes','Device Type',],inplace=True)
cleaned_data.head()
```

```python
cleaned_data.rename(columns={'Duration':'duration','Start Time':'start_time','Title':'title'},inplace=True)
```

```python
cleaned_data.info()
```

```python
cleaned_data['start_time'] = pd.to_datetime(cleaned_data['start_time'])
cleaned_data['duration'] = pd.to_timedelta(cleaned_data['duration'])
cleaned_data.info()
```

```python
cleaned_data = cleaned_data[cleaned_data['duration'] >= pd.Timedelta(minutes = 5)]
```

```python
cleaned_data = cleaned_data[cleaned_data['start_time'] >= pd.Timestamp(year=2025, month=1, day=1)]
cleaned_data
```

```python
cleaned_data['title'].unique()
```

```python
import re
def clean_title(title):
    if pd.isna(title):
        return title
    
    cleaned = re.split(r'\bSeason\b|\bEpisode\b|\bPart\b|\bLimited\b', title, flags=re.IGNORECASE)[0]
    cleaned = cleaned.strip()
    cleaned = cleaned.rstrip(':').strip()
    return cleaned
cleaned_data['title'] = cleaned_data['title'].apply(clean_title)
cleaned_data['title'].unique()
```

```python
print(cleaned_data.head())
print(cleaned_data.describe())
print(cleaned_data.info())
```

```python
cleaned_data.to_csv('../data/processed/cleaned_viewing.csv',index=False)
print("Cleaned data saved to '../data/processed/cleaned_viewing.csv'")
```
